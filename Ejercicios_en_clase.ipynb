```python
class Punto:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def determinar_cuadrante(self):
        if self.x == 0 and self.y == 0:
            print(f"El punto ({self.x}, {self.y}) está en el origen.")
        elif self.x == 0:
            print(f"El punto ({self.x}, {self.y}) está sobre el eje Y.")
        elif self.y == 0:
            print(f"El punto ({self.x}, {self.y}) está sobre el eje X.")
        elif self.x > 0 and self.y > 0:
            print(f"El punto ({self.x}, {self.y}) está en el Cuadrante I.")
        elif self.x < 0 and self.y > 0:
            print(f"El punto ({self.x}, {self.y}) está en el Cuadrante II.")
        elif self.x < 0 and self.y < 0:
            print(f"El punto ({self.x}, {self.y}) está en el Cuadrante III.")
        elif self.x > 0 and self.y < 0:
            print(f"El punto ({self.x}, {self.y}) está en el Cuadrante IV.")

# --- Ejemplos de uso de la clase Punto ---
print("\n--- Ejemplos de Puntos ---")

punto1 = Punto(3, 5)
print("Punto 1:", end=" ")
punto1.determinar_cuadrante()

punto2 = Punto(-2, 4)
print("Punto 2:", end=" ")
punto2.determinar_cuadrante()

punto3 = Punto(-1, -6)
print("Punto 3:", end=" ")
punto3.determinar_cuadrante()

punto4 = Punto(7, -2)
print("Punto 4:", end=" ")
punto4.determinar_cuadrante()

punto_origen = Punto(0, 0)
print("Punto Origen:", end=" ")
punto_origen.determinar_cuadrante()

punto_eje_x = Punto(5, 0)
print("Punto en Eje X:", end=" ")
punto_eje_x.determinar_cuadrante()

punto_eje_y = Punto(0, -3)
print("Punto en Eje Y:", end=" ")
punto_eje_y.determinar_cuadrante()
```
